```python
import pandas as pd
import os
from collections import defaultdict

df_dict = defaultdict(list)
path = "/Users/deqiwan/Downloads/3rdsemester/522 Time Series/project/temp_data"

for filename in os.listdir(path):
    if filename[:3] == ".DS":
        continue
    else:
        df_dict[filename[:3]] = pd.read_csv(os.path.join(path, filename), sep=",", header=0)
#         df_dict[filename[:3]].set_index('valid',inplace=True)
        df_dict[filename[:3]]['valid'] = pd.to_datetime(df_dict[filename[:3]]['valid'])
        df_dict[filename[:3]]['valid'] = df_dict[filename[:3]]['valid'].apply(lambda x: x.ceil(freq ='H'))
```

```python
df_dict.keys()
```

```python
# change column name -> "temp_LAX"
for k,i in df_dict.items():
    i.columns = [j+'_'+k if j != "valid" else j for j in i.columns ]
```

```python
for key, val in df_dict.items():
    print(val.shape)
```

```python
df_dict_agg = defaultdict(list)
for key, val in df_dict.items():
    df_dict_agg[key] = val.groupby('valid').agg('mean')
```

```python
for key, val in df_dict_agg.items():
    print(val.shape)
```

```python
df_dict_agg['LAX'].columns
```

```python
df = pd.merge(df_dict_agg["RIV"], df_dict_agg["TRM"], on="valid", how='inner')
df = pd.merge(df, df_dict_agg["WJF"], on="valid", how='inner')
df = pd.merge(df, df_dict_agg["CQT"], on="valid", how='inner')
df = pd.merge(df, df_dict_agg["LAX"], on="valid", how='inner')
```
